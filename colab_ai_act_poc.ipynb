# EU AI Act – Colab POC
Upload the Excel workbook and preview the key sheets.


```python
import pandas as pd
from google.colab import files
from io import BytesIO
from ai_act_loader import load_workbook, validate_sheets, load_core_frames

print('Upload the EU-AI-ACT_Playbook.xlsx or EU-AI-ACT_Playbook_1.xlsx')
uploaded = files.upload()
fname = next(iter(uploaded.keys()))
xls = load_workbook(BytesIO(uploaded[fname]))
print('Sheets found:', xls.sheet_names)
missing = validate_sheets(xls)
if missing:
    print('WARNING – Missing sheets:', missing)
frames = load_core_frames(xls)
for k, df in frames.items():
    print('\n===', k, '===')
    display(df.head(10))
```
